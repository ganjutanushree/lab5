```python
# 1.Loading the DataSet
from sklearn.datasets import fetch_california_housing
import pandas as pd  

california = fetch_california_housing()  #california dataset

tanu= pd.DataFrame(california.data, columns=california.feature_names)   #convert to pandas dataframe

tanu['MedHouseVal'] = california.target  #add target

print("First 5 rows of the California Housing dataset:")     #display 5 rows
print(tanu.head())

print("\nSummary statistics:")    #print Summary statistics
print(tanu.describe())
```

```python
#2.Visual Data analysis
import matplotlib.pyplot as plt

plt.hist(tanu['MedHouseVal'], bins=43)       #histogram of MedHouseVal
plt.title('MedHouseVal Distribution')
plt.xlabel('MedHouseVal')
plt.ylabel('Frequency')
plt.show()

plt.scatter(tanu['MedInc'], tanu['MedHouseVal'])   # Scatter plot -- MedInc vs MedHouseVal
plt.title(' MedHouseVal vs  MedInc ')
plt.xlabel('MedInc')
plt.ylabel('MedHouseVal')
plt.show()

plt.scatter(tanu['AveRooms'], tanu['MedHouseVal'])     #AveRooms vs MedHouseVal
plt.title('MedHouseVal vs AveRooms')
plt.xlabel('AveRooms')
plt.ylabel('MedHouseVal')
plt.show()
```
